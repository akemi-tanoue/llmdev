仕様
英単語の入力を受け付けます。未入力でリターン（エンター）キーを押されるまで、英単語の受付を繰り返します。受付が完了したら、入力された英単語をABC順に並べ替えて表示します。また、すべての英単語に含まれるアルファベットごとの数も表示します。

解き方のヒント
いきなりソースコードを書くのではなく、プログラム思考で仕様を整理しましょう。考え方の例を示します。

英単語を格納するための空のリストを作成する
アルファベットごとの数を格納する辞書を作成する。キーがアルファベット小文字で、値が0の辞書を作成する
'a':0, 'b':0, 'c':0, 'd':0, 'e':0, 'f':0, 'g':0, 'h':0, 'i':0, 'j':0, 'k':0, 'l':0, 'm':0, 'n':0, 'o':0, 'p':0, 'q':0, 'r':0, 's':0, 't':0, 'u':0, 'v':0, 'w':0, 'x':0, 'y':0, 'z':0
英単語の入力の受付を繰り返す
キーボードから英単語の入力を受け付ける
入力された文字が "" なら繰り返しを抜ける
英単語をリストに格納する
英単語を構成する文字を1文字ずつ確認して、登場したアルファベットの数に1を加算する
英単語のリストをABC順に並べ替える
英単語のリストを一覧表示する
アルファベットごとの数を格納する辞書から、数が0より大きい文字について「○が△個ありました」と表示する

```python
#英単語を格納するための空のリストを作成する
words = []

# アルファベットごとの数を格納する辞書を作成する。キーがアルファベット小文字で、値が0の辞書を作成する
# 'a':0, 'b':0, 'c':0, 'd':0, 'e':0, 'f':0, 'g':0, 'h':0, 'i':0, 'j':0, 'k':0, 'l':0, 'm':0, 'n':0, 'o':0, 'p':0, 'q':0, 'r':0, 's':0, 't':0, 'u':0, 'v':0, 'w':0, 'x':0, 'y':0, 'z':0
char_dic = dict.fromkeys(list("abcdefghijklmnopqrstuvwxyz"), 0)

#キーボードから英単語の入力を受け付ける
#英単語の入力の受付を繰り返す
#キーボードから英単語の入力を受け付ける
#入力された文字が "" なら繰り返しを抜ける
#英単語をリストに格納する
#英単語を構成する文字を1文字ずつ確認して、登場したアルファベットの数に1を加算する
while True:
  word = input("英単語を入力してください：")
  if word == "":
    break  # ループを抜ける
  words.append(word)  # 入力された英単語をリストに追加
  for char in word:
    char_dic[char] += 1

#英単語のリストをABC順に並べ替える
#英単語のリストを一覧表示する

words.sort()
print("入力した英単語：", words) #並び替えたものを出力する

#アルファベットごとの数を格納する辞書から、数が0より大きい文字について「○が△個ありました」と表示する
for char, count in char_dic.items():
  if count > 0:
    print(f"「{char}」が{count}個ありました")
```
